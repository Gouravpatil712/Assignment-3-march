## Question 01 - Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

# Answer :-

Seaborn is a popular Python library that is built on top of Matplotlib and is used for creating visually appealing statistical graphs. Some of the common types of plots that can be created using Seaborn are:

1. Scatter plots: A scatter plot is used to visualize the relationship between two numerical variables. Seaborn's scatter plot allows us to visualize this relationship using colors and markers that can represent additional variables.

2. Line plots: Line plots are used to visualize the trend of a numerical variable over time. Seaborn's line plot allows us to add multiple lines on the same plot to compare trends across multiple variables.

3. Bar plots: Bar plots are used to visualize the distribution of a categorical variable. Seaborn's bar plot allows us to show the mean and confidence interval of the data.

4. Heat maps: Heat maps are used to visualize the relationship between two categorical variables. Seaborn's heat map allows us to visualize the data using colors and is often used for data analysis and feature selection.

5. Violin plots: Violin plots are used to visualize the distribution of a numerical variable across multiple categories. Seaborn's violin plot is similar to a box plot, but it also shows the density of the data at different values.

Overall, Seaborn is a powerful tool for data visualization and provides many options for creating visually appealing plots that can help us better understand our data.

## Question 02 - Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x = "timepoint" and y = "signal" for different events and regions.

```python
# Answer :-

import seaborn as sns

# Load the "fmri" dataset
fmri = sns.load_dataset("fmri")
```

```python
fmri.head()
```

```python
# Plot a line plot using x = "timepoint" and y = "signal" for different events and regions
sns.lineplot(x=fmri["timepoint"], y=fmri["signal"], hue=fmri["event"], style=fmri["region"])
```

In this code, we first load the "fmri" dataset using the load_dataset function of seaborn. Then, we use the lineplot function to plot a line plot using x=fmri["timepoint"] and y=fmri["signal"]. We use the hue=fmri["event"] parameter to group the data by the "event" column and use different colors for each group. We also use the style=fmri["region"] parameter to differentiate the data by the "region" column and use different line styles for each group.

## Question 03 - Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x = 'pclass', y = 'age' and y = 'fare'.
Note: pclass, age, and fare are columns in the titanic dataset.

```python
# Answer :- 

import seaborn as sns

# Load titanic dataset
titanic = sns.load_dataset("titanic")
```

```python
titanic.head()
```

```python
# Plot box plot for 'pclass' vs 'age'
sns.boxplot(x='pclass', y='age', data=titanic)
```

```python
# Plot box plot for 'fare' vs 'age'
sns.boxplot(x='pclass', y='fare', data=titanic)
```

The first box plot shows the distribution of ages across different passenger classes, and the second box plot shows the distribution of fares for different passenger classes.

## Question 04 - Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue parameter for the 'cut' column of the diamonds dataset.

```python
## Answer :-

import seaborn as sns

# load the diamonds dataset
diamonds = sns.load_dataset('diamonds')
```

```python
diamonds.head()
```

```python
# plot a histogram of the 'price' column with the 'cut' column as the hue
sns.histplot(data=diamonds, x='price', hue='cut', multiple='stack')
```

This code first loads the "diamonds" dataset using the sns.load_dataset() function from seaborn. It then uses the sns.histplot() function to plot a histogram of the 'price' column, with the 'cut' column as the hue parameter. The multiple='stack' parameter tells seaborn to stack the histograms for each value of the 'cut' column.

This will generate a histogram that shows the distribution of prices for each level of cut. You can customize the appearance of the plot using the many available parameters of the sns.histplot() function.

## Question 06 - Use the "flights" dataset from seaborn to plot a heatmap.

```python
# Answer :-

import seaborn as sns
import matplotlib.pyplot as plt

# Load the flights dataset
flights = sns.load_dataset('flights')
```

```python
flights.head()
```

```python
# Pivot the data to create a matrix of monthly passenger counts
flights_matrix = flights.pivot('month', 'year', 'passengers')

# Create a heatmap of the monthly passenger counts
sns.heatmap(flights_matrix, cmap='Blues')

# Set the plot title
plt.title('Monthly Passenger Counts (1949-1960)')

# Show the plot
plt.show()
```
